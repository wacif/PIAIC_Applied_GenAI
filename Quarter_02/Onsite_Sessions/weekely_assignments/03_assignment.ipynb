```python
# Import the genai module for API interaction
import google.generativeai as genai
from google.colab import userdata
```

```python
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')

# Configure it with your api key
genai.configure(api_key=GEMINI_API_KEY)
```

```python
# Display the downloaded image
import PIL.Image
img = PIL.Image.open('/content/handguns.jpg')
display(img)
```

```python
# Write a prompt and store it in some variable
prompt = """1- what you see in the picture?
2- How many gund are there in the picture? also named them.
Your answer must be in markdown format."""
```

```python
model = genai.GenerativeModel("gemini-2.0-flash-exp")
response = model.generate_content([prompt, img])
```

```python
from IPython.display import Markdown
display(Markdown(response.text))
```
